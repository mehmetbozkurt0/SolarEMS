```python
import pandas as pd
import numpy as np

# NOT: Bu kodun Ã§alÄ±ÅŸmasÄ± iÃ§in Ã¶nceki adÄ±mlardaki 'test_data' ve 'model_tuketim' 
# deÄŸiÅŸkenlerinin hafÄ±zada olmasÄ± veya tekrar yÃ¼klenmesi gerekir.
# Veriyi tekrar hazÄ±rlayalÄ±m (Garanti olsun):
print("Veriler ve ModÃ¼ller HazÄ±rlanÄ±yor...")

# Ã–nceki verileri hÄ±zlÄ±ca yÃ¼kleyip hazÄ±rlayalÄ±m
df_solar = pd.read_csv('temiz_solar_veri.csv', index_col='time', parse_dates=True).resample('h').mean()
df_tuketim = pd.read_csv('temiz_tuketim_veri_5000kWh.csv', index_col='dt', parse_dates=True).resample('h').mean()
# Timezone temizliÄŸi
try:
    df_solar.index = df_solar.index.tz_localize(None)
    df_tuketim.index = df_tuketim.index.tz_localize(None)
except: pass

# Test iÃ§in ortak veriyi birleÅŸtir
df_final = pd.concat([df_tuketim, df_solar[['Uretim_W', 'Sicaklik_C']]], axis=1, join='inner').dropna()

# SimÃ¼lasyon: GerÃ§ek Ã¼retim, Tahmin edilenden bazen dÃ¼ÅŸÃ¼k olsun (ArÄ±za SimÃ¼lasyonu)
# Normalde 'Tahmin'i modelden alÄ±rÄ±z, burada simÃ¼le ediyoruz.
df_final['Tahmin_Uretim_W'] = df_final['Uretim_W'] * 1.05 # Model genelde %5 fazla bekliyor olsun
# Rastgele bazÄ± saatlerde paneli bozalÄ±m (Verimi %40'a dÃ¼ÅŸÃ¼relim)
np.random.seed(101)
ariza_indeksleri = np.random.choice(df_final.index, size=20, replace=False)
df_final.loc[ariza_indeksleri, 'Uretim_W'] = df_final.loc[ariza_indeksleri, 'Uretim_W'] * 0.4 


# =============================================================================
# MODÃœL 1: ANOMALÄ° (ARIZA) TESPÄ°T SÄ°STEMÄ° (Ä°ÅžL.11)
# =============================================================================
print("\nðŸ” MODÃœL 1: Anomali Tespiti Ã‡alÄ±ÅŸtÄ±rÄ±lÄ±yor...")

# Kural: EÄŸer GerÃ§ek Ãœretim, Tahmin edilenin %60'Ä±ndan bile azsa -> ARIZA VARDIR!
# (Hava kapalÄ± olabilir ama model bunu zaten bilir. Bu fark 'beklenmeyen' farktÄ±r.)
df_final['Performans_Orani'] = df_final['Uretim_W'] / df_final['Tahmin_Uretim_W']
df_final['Anomali_Var_Mi'] = (df_final['Performans_Orani'] < 0.60) & (df_final['Tahmin_Uretim_W'] > 100) # Geceleri hariÃ§ tut

anomaliler = df_final[df_final['Anomali_Var_Mi']]

print(f"   -> Toplam Taranan Saat: {len(df_final)}")
print(f"   -> Tespit Edilen Kritik ArÄ±za/GÃ¶lgeleme SayÄ±sÄ±: {len(anomaliler)}")

if not anomaliler.empty:
    ornek_an = anomaliler.index[0]
    print(f"   âš ï¸ Ã–RNEK UYARI ({ornek_an}):")
    print(f"      Beklenen: {df_final.loc[ornek_an, 'Tahmin_Uretim_W']:.0f} W")
    print(f"      GerÃ§ekleÅŸen: {df_final.loc[ornek_an, 'Uretim_W']:.0f} W (Performans OranÄ± %{df_final.loc[ornek_an, 'Performans_Orani']*100:.0f})")
    print("      SÄ°STEM MESAJI: 'Panelde beklenmedik verim dÃ¼ÅŸÃ¼ÅŸÃ¼! Kirlenme veya gÃ¶lgeleme kontrolÃ¼ yapÄ±n.'")

# =============================================================================
# MODÃœL 2: KARBON AYAK Ä°ZÄ° HESAPLAYICI (Ä°ÅžL.07)
# =============================================================================
print("\nðŸŒ± MODÃœL 2: Ã‡evresel Etki (Karbon Ayak Ä°zi) HesaplanÄ±yor...")

# KatsayÄ±lar (TÃ¼rkiye Åžebeke OrtalamasÄ±)
# 1 kWh Elektrik Ãœretimi = ~0.44 kg CO2 salÄ±nÄ±mÄ± (DoÄŸalgaz/KÃ¶mÃ¼r karÄ±ÅŸÄ±k)
# 1 YetiÅŸkin AÄŸaÃ§ = YÄ±lda ortalama 20 kg CO2 emer.
CO2_KATSAYISI = 0.44 # kg/kWh

toplam_uretim_kwh = df_final['Uretim_W'].sum() / 1000
onlenen_co2_kg = toplam_uretim_kwh * CO2_KATSAYISI
kurtarilan_agac = onlenen_co2_kg / 20

print(f"   -> Toplam Temiz Enerji Ãœretimi: {toplam_uretim_kwh:.2f} kWh")
print(f"   -> Ã–nlenen Karbon SalÄ±nÄ±mÄ± (CO2): {onlenen_co2_kg:.2f} kg")
print(f"   -> ðŸŒ³ EÅžDEÄžER DOÄžA KATKISI: {kurtarilan_agac:.1f} adet aÄŸaÃ§ dikmiÅŸ kadar fayda saÄŸladÄ±nÄ±z!")
```
